```
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```
df = pd.read_csv("praticas_inovadoras_obama.csv")
```

```
df.head()
```

```
colunas_texto = ['id','campo_atuacao', 'quais_recursos_ja_utilizava', 'quais_recursos_passou_utilizar', 
                 'Após a formação, você utilizou/utiliza algum dos recursos e ensinamentos passados?',
                 'tres_caracteristicas_aula', 'praticas_inovadoras_oqeh']
```

```
df_numerico = df.drop(columns = colunas_texto)
```

```
corr = df_numerico.corr(method='pearson')
mask = np.triu(np.ones_like(corr, dtype=np.bool))
```

```
f, ax = plt.subplots(figsize=(15, 13))
cmap = sns.diverging_palette(220, 10, as_cmap=True)
sns.heatmap(corr, mask=mask, cmap=cmap, annot=True, fmt=".1", vmax=.3, center=0,
            square=True, linewidths=.5, cbar_kws={"shrink": .5})
```
